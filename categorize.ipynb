將須完成MBOM表所要input的檔案分成以下6個種類，使各自種類有特定的邏輯去判斷擷取條件

```python
import os
import shutil
import re

#1.1開始
# 為要分類的PDF檔案所在路徑
base_path = r"G:\共用雲端硬碟\SongLab2024畢_奕傑、顓宏、凱威\奕傑\碩士論文相關\碩士_Coding"

# 定義類別對應的資料夾名稱及匹配順序
categories = [
    (r'PSE_[0-9][A-Za-z0-9]*_PL', '_PL'),  # PSE_ 後面第一個字是數字的情況
    (r'PSE_[A-Za-z][A-Za-z0-9]*_PL', '-X_PL'),  # PSE_ 後面第一個字是英文的情況
    (r'-[A-Za-z0-9]+_PL', '-X_PL'),  # 保留原來的 -X_PL 判斷邏輯
    ('EPL', '_EPL'),
    ('MPL', '-X_PL'),
    ('MPPL', '-X_PL'),
    ('xml', '_xml'),
    ('BE-PL', '_BE-PL')
]

# 新增八種資料夾並在每個資料夾中創建 pdf_file、pdf2txt、final_output 資料夾
for _, folder_name in categories:
    category_folder = os.path.join(base_path, folder_name)
    os.makedirs(category_folder, exist_ok=True)
    pdf_file_folder = os.path.join(category_folder, 'pdf_file')
    pdf2txt_folder = os.path.join(category_folder, 'pdf2txt')
    final_output_folder = os.path.join(category_folder, 'final_output')
    os.makedirs(pdf_file_folder, exist_ok=True)
    os.makedirs(pdf2txt_folder, exist_ok=True)
    os.makedirs(final_output_folder, exist_ok=True)
    print(f'資料夾 {category_folder} 及其子資料夾建立成功！')

# 新增 "others" 資料夾
others_folder = os.path.join(base_path, 'others')
os.makedirs(others_folder, exist_ok=True)
print(f'資料夾 {others_folder} 建立成功！')
#1.1結束
#--------------------------------------------------------------------------------------------------------------
#1.2開始
# 取得所有PDF檔案
pdf_files = [f for f in os.listdir(base_path) if f.endswith('.pdf')]

# 分類PDF檔案並移動到對應的 pdf_file 資料夾中
for pdf_file in pdf_files:
    is_matched = False
    
    # 檢查是否符合任一分類條件
    for pattern, folder_name in categories:
        if re.search(pattern, pdf_file):
            source_file = os.path.join(base_path, pdf_file)
            destination_folder = os.path.join(base_path, folder_name, 'pdf_file')
            shutil.move(source_file, destination_folder)
            print(f'移動檔案 {pdf_file} 到資料夾 {destination_folder}')
            is_matched = True
            break
    
    # 如果沒有符合任何分類條件，再檢查是否只包含數字和英文字母且無標點符號，有的話將該檔案放到-X_PL的資料夾中
    if not is_matched and pdf_file[:-4].isalnum() and not re.search(r'[-_]', pdf_file[:-4]):
        source_file = os.path.join(base_path, pdf_file)
        destination_folder = os.path.join(base_path, '-X_PL', 'pdf_file')
        shutil.move(source_file, destination_folder)
        print(f'移動檔案 {pdf_file} 到資料夾 {destination_folder}')
        is_matched = True
    
    # 如果沒有符合任何分類條件，移動到 others 資料夾
    if not is_matched:
        source_file = os.path.join(base_path, pdf_file)
        destination_folder = os.path.join(base_path, 'others')
        shutil.move(source_file, destination_folder)
        print(f'移動檔案 {pdf_file} 到資料夾 {destination_folder}')
#1.2結束
print('分類完成！')
```
